```python
import numpy as np
import matplotlib.pyplot as plt

# Define the initial weights and learning rate
W0 = 10
W1 = 0.2
W2 = -0.75
learning_rate = 0.05

# Define the AND gate input data and target values
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 0, 0, 1])

# Initialize the error list to store error values for each epoch
error_list = []

# Define the step activation function
def step_function(x):
    return 1 if x >= 0 else 0

# Training the perceptron
epochs = 0
converged = False

while not converged:
    total_error = 0
    for i in range(len(X)):
        # Calculate the output of the perceptron
        net = W0 + W1 * X[i, 0] + W2 * X[i, 1]
        output = step_function(net)
        
        # Calculate the error and update weights
        error = Y[i] - output
        total_error += error ** 2
        W0 += learning_rate * error
        W1 += learning_rate * error * X[i, 0]
        W2 += learning_rate * error * X[i, 1]
    
        # Append the total error for this epoch to the error list
    error_list.append(total_error)
    
    # Check for convergence
    if total_error == 0:
        converged = True
    
    epochs += 1

# Plotting the error values
plt.plot(range(epochs), error_list)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error vs. Epochs')
plt.show()

# Print the number of epochs needed for convergence
print(f'Number of epochs needed for convergence: {epochs}')

# Print the final weights
print(f'Final weights: W0={W0}, W1={W1}, W2={W2}')
```
